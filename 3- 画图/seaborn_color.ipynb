```python
%matplotlib inline
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize':(6,6)})
np.random.seed(sum(map(ord,'palette')))
```

### 建立调色板
- qualitative
- sequential
- diverging

```python
# 查看调色板
# 这个函数的返回值总是一个由RGB元组组成的列表，无参数调用color_palette函数则会返回当前默认的色环的列表。
sns.color_palette()
```

#### - 分类调色板

```python
current_palette = sns.color_palette()
sns.palplot(current_palette)
```

```python
# 分类调色板有6中不同的主题
themes = ['deep','muted','pastel','bright','dark','colorblind']
for theme in themes:
    current_palette = sns.color_palette(theme)
    sns.palplot(current_palette)
```

#### cubehelix 连续调色板
cubehelix调色板系统在亮度和色变变化上具有线性上升或下降的特点。

```python
sns.palplot(sns.color_palette('cubehelix',8))
```

```python
sns.palplot(sns.cubehelix_palette(8))
```

```python
sns.palplot(sns.cubehelix_palette(8,start=.5,rot=-.75))
```

```python
#  等高线例子
x, y = np.random.multivariate_normal([0,0],[[1,-0.5],[-0.5,1]] ,size=300).T
cmap = sns.cubehelix_palette(light=1,as_cmap=True)
sns.kdeplot(x,y,cmap=cmap,shade=True)
```

### 离散色板
调色板中的第三类被称为“离散”。这类色板适用于数据特征含有大的低值和大的高值。数据中通常有一个意义明确的中点。

```python
sns.palplot(sns.diverging_palette(220,20,n=7))
```

```python
sns.palplot(sns.diverging_palette(145,280,s=85,l=25,n=7))
```

```python
# 设置默认的调色板
def sinplot(flip=1):
    x = np.linspace(0,14,100)
    for i in range(1,7):
        plt.plot(x,np.sin(x+i*5)*(7-i)*flip)
sns.set_palette('husl')
sinplot()
```

```python
# 局域调色板
with sns.color_palette('BuGn_d'):
    sinplot()
```
