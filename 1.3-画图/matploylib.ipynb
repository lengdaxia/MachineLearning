# 导入数据集


```python
import pandas as pd

# 美国失业率
unrate =  pd.read_csv('UNRATE.csv')
unrate['DATE'] = pd.to_datetime(unrate['DATE'])

print(unrate.head())
```

```python
import matplotlib.pyplot as plt
#%matplotlib inline

# 空白画布
plt.plot()
plt.show()
```

## 折线图
---

```python
first = unrate[0:12]

plt.plot(first['DATE'],first['VALUE'])
# title
plt.title('Monthly Unemployment rate')
# xlalabel
plt.xlabel('Month')
# ylabel
plt.ylabel('unrate ratio')

# x 旋转45度
plt.xticks(rotation=45)
plt.show()
```

## 子图绘制
---

```python
#创建 2x2个子图 
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,4)
# ax4 = fig.add_subplot(2,2,3)
plt.show()
```

```python
import numpy  as np

fig = plt.figure(figsize=(8,8))

ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

ax1.plot(np.random.randint(1,5,5),np.arange(5))
ax2.plot(np.arange(10)*3,np.arange(10))

plt.show()
```

```python
# 在同一个图中画出两条折线
unrate['MONTH'] = unrate['DATE'].dt.month

fig = plt.figure(figsize=(12,6))

plt.plot(unrate[0:12]['MONTH'],unrate[0:12]['VALUE'],c='red')                 
plt.plot(unrate[12:24]['MONTH'],unrate[12:24]['VALUE'],c='blue')

plt.show()
```

```python
# 多条折线图并展示线条信息
fig = plt.figure(figsize=(12,6))
colors = ['red','blue','orange','gray','green']

for i in range(len(colors)):
    start_idx = i * 12
    end_idx = (i+1)*12
    subset = unrate[start_idx:end_idx]
    label = str(1948+i)
    plt.plot(subset['MONTH'],subset['VALUE'],c=colors[i],label = label)
# plt.legend(loc='upper left')
plt.legend()
plt.ylabel('ratio')
plt.xlabel('month')
plt.title('America Unemployment ratio')
plt.show()
```

## 柱形图
---

```python
# 导入数据

reviews = pd.read_csv('fandango_scores.csv')

cols = ['FILM','RT_user_norm','Metacritic_user_nom','IMDB_norm','Fandango_Ratingvalue','Fandango_Stars']
norms_reviews = reviews[cols]
print(norms_reviews[:1])
```

```python
# 柱形图
import matplotlib.pyplot as plt
import numpy as np

num_cols = ['FILM','RT_user_norm','Metacritic_user_nom','IMDB_norm','Fandango_Ratingvalue','Fandango_Stars']
bar_height = norms_reviews[:1][num_cols].values[0][1:6]
print(bar_height)
bar_position = np.arange(5) + 0.75
print(bar_position)

fig, ax = plt.subplots()
ax.bar(bar_position,bar_height,0.5)
plt.show()
```

## 散点图
---

```python
fig, ax = plt.subplots()
ax.scatter(norms_reviews['Fandango_Ratingvalue'],norms_reviews['RT_user_norm'])

ax.set_xlabel('Fandango')
ax.set_ylabel('Totten tomatos')
plt.show()
```

## 柱形图和盒图

```python
# 柱形图

fig, ax = plt.subplots()

ax.hist(norms_reviews['Fandango_Ratingvalue'],bins=20,range=(4,5))

plt.show()
```

```python
# 盒图
fig ,ax = plt.subplots()

ax.boxplot(norms_reviews['Fandango_Ratingvalue'])
plt.show()
```

```python

# 多个盒图
fig, ax = plt.subplots()
cols = ['RT_user_norm','Metacritic_user_nom','IMDB_norm','Fandango_Ratingvalue','Fandango_Stars']
ax.boxplot(norms_reviews[cols].values)
plt.show()
```
